###Biểu đồ thể hiện biến động doanh số theo tháng

```python
#biểu đồ thể hiện biến động doanh số theo tháng
plt.figure(figsize=(20,15))
df['month'].value_counts().plot.pie()
```

### Doanh thu trung bình trong 3 tháng gần nhất

```python
#Doanh thu trung bình trong 3 tháng gần nhất
month_sales = df.groupby(['month']).sum().sort_values('year',ascending=False).head(3)
month_sales.reset_index(inplace=True)
month_sales
```

### Biểu đồ doanh số bán hàng theo quý

```python
#Biểu đồ doanh số bán hàng theo quý
plt.figure(figsize=(20,15))
df.groupby(df['Order Date'].dt.to_period('Q'))['Sales'].sum().plot.pie()
```

### Quý có tổng doanh số cao nhất

```python
#Quý có tổng doanh số cao nhất
top_Q=df.groupby(df['Order Date'].dt.to_period('Q'))['Sales'].sum().sort_values(ascending=False).head(1)
top_Q
```

### Biểu đồ thể hiện biến động thời gian giao hàng theo từng tháng

```python
#Tính thời gian giao hàng trung bình theo từng tháng và vẽ biểu đồ
df['duration'] = abs(df['Ship Date']-df['Order Date'])
df_dura=df.groupby(['month'])['duration'].sum().reset_index()
fig, ax = plt.subplots()
ax.plot(df_dura['month'], df_dura['duration'])

ax.set(xlabel='Month', ylabel='Duration',
       title='Thời gian giao hàng trung bình theo tháng')
ax.grid()

plt.show()
```

### Biểu đồ thể hiện sức mua theo từng Thành phố

```python
# Sức mua theo từng Thành phố
plt.figure(figsize=(20,15))
#df['City'].value_counts().plot.pie()
df.groupby(['City'])['Sales'].sum().plot.pie()
```

### Danh sách Top 3 thành phố có doanh số thấp nhất

```python
#Top 3 thành phố có doanh số thấp nhất
top_city_min=df.groupby(['City']).sum().sort_values('Sales',ascending=True).head(3)
top_city_min
```

### Danh sách Top 3 thành phố có doanh số cao nhất

```python
#Top 3 thành phố có doanh số cao nhất
top_city_max=df.groupby(['City']).sum().sort_values('Sales',ascending=False).head(3)
top_city_max
```

### Biểu đồ tỉ trọng doanh số theo bang

```python
#tỉ trọng doanh số theo bang
df_tongsales=df['Sales'].sum()
df_chia=df.groupby(['State'])['Sales'].sum().divide(other = df_tongsales)
kq=df_chia*100
plt.figure(figsize=(20,15))
kq.plot.pie()
```

### Trung bình doanh số của 5 bang dẫn đầu về sức mua

```python
#trung bình doanh số của 5 bang dẫn đầu về sức mua
top_states_aver=df.groupby(['State']).mean().sort_values('Sales',ascending=False).head(5)
top_states_aver=top_states_aver[['Sales']].round(2)
top_states_aver.reset_index(inplace=True)
top_states_aver
```

### Danh sách 5 bang dẫn đầu doanh số theo từng vùng

```python
#5 bang dẫn đầu doanh số theo từng vùng
top_region=df.groupby(['Region']).sum().sort_values('Sales',ascending=False).head(5)
top_region=top_region[['Sales']].round(2)
top_region.reset_index(inplace=True)
top_region
```

### Danh sách Top 3 sản phẩm có doanh số cao nhất

```python
#Top 3 sản phẩm có doanh số cao nhất
top_product=df.groupby(['Product Name']).sum().sort_values('Sales',ascending=False).head(3)
top_product=top_product[['Sales']].round(2)
top_product.reset_index(inplace=True)
top_product
```

### Biểu đồ tỉ trọng doanh số theo từng category

```python
#Tỉ trọng doanh số theo từng category
df_tongsales=df['Sales'].sum()
df_chia=df.groupby(['Category'])['Sales'].sum().divide(other = df_tongsales)
kq_ca=df_chia*100
plt.figure(figsize=(20,15))
kq_ca.plot.pie()
```
